```python
import numpy

import math

from matplotlib import pyplot

%matplotlib inline
```

```python
N = 51

x_start, x_end = -2.0, 2.0

y_start, y_end = -0.5, 0.5

x = numpy.linspace(x_start, x_end, N)

y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x, y)
```

```python
numpy.shape(X)
```

```python
N1 = 11
gamma = numpy.ones(N1)*1
x_vortex = numpy.linspace(-2.0, 2.0, N1)
y_vortex = numpy.zeros(N1)
```

```python
numpy.shape(gamma)
```

```python
numpy.shape(x_vortex)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
```

Superposition

```python
u_vortex1 = 0.0
v_vortex1 = 0.0


for i in range(N1):
    u_vortex, v_vortex = get_velocity_vortex(gamma[i], x_vortex[i], y_vortex[i], X, Y)
    
    u_vortex1 += u_vortex
    v_vortex1 += v_vortex
```

```python
size = 10

pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))

pyplot.grid(True)

pyplot.xlabel('x', fontsize=16)

pyplot.ylabel('y', fontsize=16)

pyplot.xlim(-1.0, 1.0)

pyplot.ylim(-0.25, 0.25)

pyplot.streamplot(X, Y, u_vortex1, v_vortex1, density=4, linewidth=1, arrowsize=1, arrowstyle='->'
            )
pyplot.scatter(x_vortex, y_vortex, color='r', s=60, marker='o')
```
